```python
df['Last Updated'].value_counts(['Category', 'Year'], sort=False)
```

```python
# Sum up count of last updated for year and category
df1 = df.groupby(['Category','Year'])['Year'].count().reset_index(name="Count")
# df1.to_frame()
df1
```

```python
df1.columns
```

```python
final_df = df1.pivot(index='Category', columns='Year', values='Count')
final_df
```

```python
final_df.columns
```

```python
final_df = final_df[[' 2017', ' 2018']]
curr = final_df[' 2018']
prev = final_df[' 2017']
final_df['Popularity'] = (curr - prev) / prev
final_df
```

```python
final_df.to_csv('popularity.csv')
```
